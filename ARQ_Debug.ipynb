```python
from pymavlink import mavutil
import datetime
import time

fcLink = mavutil.mavlink_connection('COM7', baud=57600)
rcLink = mavutil.mavlink_connection('COM4', baud=230400)

time.sleep(1)  #Give some time for serial to wake up

startTime = datetime.datetime.now()

while (datetime.datetime.now() - startTime).total_seconds() < 2.0:
    fcMsg = fcLink.recv_match(blocking=True)
    fcMsgType = fcMsg.get_type()
    fcMsgSeq = fcMsg.get_seq()

    if fcMsgSeq > 0:
        print(f'Time: {(datetime.datetime.now() - startTime).total_seconds()} | FC Message: {fcMsgType} | Sequence Number: {fcMsgSeq}')

    rcMsg = rcLink.recv_match(blocking=True)
    rcMsgType = rcMsg.get_type()
    rcMsgSeq = rcMsg.get_seq()

    if rcMsgSeq > 0:
        print(f'Time: {(datetime.datetime.now() - startTime).total_seconds()} | RC Message: {rcMsgType} | Sequence Number: {rcMsgSeq}')
```
